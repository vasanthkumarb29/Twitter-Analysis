**Series plot** :Since the dataset contains time-series data, we can create a line plot of the 'Close' column against the 'Date' column. This will give us an idea of how the closing prices have changed over time.

```python
plt.figure(figsize=(10,5))
plt.plot(stock_market_df['Date'], stock_market_df['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Closing Price of Twitter Inc over Time')
plt.show()
```

 **Correlation heatmap:** We can create a correlation heatmap to see the correlation between different numeric columns in the dataset.

```python
sns.heatmap(stock_market_df.corr(), cmap='coolwarm', annot=True)
plt.title('Correlation Matrix')
plt.show()
```

 **Scatter plot:** We can create a scatter plot of the 'High' column against the 'Low' column to see the relationship between the highest and lowest recorded price for the day.

```python
colors = np.random.rand(2238)

plt.scatter(stock_market_df['Date'], stock_market_df['Open'],c=colors , alpha=0.5)
plt.title('Opening price of Twitter Inc')
plt.xlabel('Date')
plt.ylabel('Open')
plt.show()
```

**Bar plot:** A bar plot is a useful visualization for showing the relationship between categorical and numerical data

```python
stock_market_df['Year'] = pd.DatetimeIndex(stock_market_df['Date']).year
df_grouped = stock_market_df.groupby('Year')['Close'].mean()
```

```python
plt.bar(df_grouped.index, df_grouped.values)
plt.xlabel('Year')
plt.ylabel('Average Closing Price')
plt.title('Average Closing Price by Year')
plt.show()
```

 **Violin plot:** We can create a violin plot of the 'Adj Close' column to see the distribution of the modified closing price based on corporate actions.

```python
sns.violinplot(x=stock_market_df['Adj Close'])
plt.xlabel('Adj Close')
plt.title('Violin Plot of Adj Close')
plt.show()
```

**Histograms:** We can create histograms of all the numeric columns in the dataset using a for loop.

```python
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']


for col in numeric_cols:
    plt.hist(stock_market_df[col],edgecolor='black')
    mean_val = np.mean(stock_market_df[col])
    median_val = np.median(stock_market_df[col])
    plt.axvline(mean_val, color='red', linewidth=1)
    plt.axvline(median_val, color='green',  linewidth=1)
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title('Histogram of '+ col)
    plt.show()
```
